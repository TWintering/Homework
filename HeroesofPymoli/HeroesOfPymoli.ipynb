```python
import pandas as pd
import numpy as np
```

```python
file="Resources/purchase_data.csv"
df=pd.read_csv(file)
```

```python
total_players=pd.DataFrame({'Total Players':df['SN'].nunique()}, index=[''])
display(total_players)
```

```python
purchasing_analysis=pd.DataFrame({'Number of Unique Items':df['Item ID'].nunique(),'Average Price':round(df['Price'].mean(),2),'Number of Purchases':df['Purchase ID'].count(),'Total Revenue':df['Price'].sum()},index=[''])
with pd.option_context('display.float_format', '${:,.2f}'.format):
    display(purchasing_analysis)
```

```python
unique=df.drop_duplicates(subset='SN')
men=unique.loc[unique['Gender']=='Male']
women=unique.loc[unique['Gender']=='Female']
other=unique.loc[(unique['Gender']!='Male') & (unique['Gender']!='Female')]
gender_demographics=pd.DataFrame({'Total Count':[men['SN'].count(),women['SN'].count(),other['SN'].count()],'Percentage of Players':[round((men['SN'].count()/unique['SN'].count())*100,2).astype(str)+'%',round((women['SN'].count()/unique['SN'].count())*100,2).astype(str)+'%',round((other['SN'].count()/unique['SN'].count())*100,2).astype(str)+'%']},index=['Male','Female','Other/Non-Disclosed'])
display(gender_demographics)
```

```python
gender=df.groupby(df['Gender'])
gender_analysis=pd.DataFrame({'Purchase Count':gender['Purchase ID'].count(),'Average Purchase Price':gender['Price'].mean(),'Total Purchase Value':gender['Price'].sum(),'Avg Total Purchase per Person':gender['Price'].sum()/gender['SN'].nunique()})
with pd.option_context('display.float_format', '${:,.2f}'.format):
    display(gender_analysis)
```

```python
bins=[0,9,14,19,24,29,34,39,999]
bin_names=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
age=unique.groupby(pd.cut(unique['Age'],bins,labels=bin_names))
age_demographics=pd.DataFrame({'Total Count':age['Purchase ID'].count(),'Percentage of Players':(round((age['Purchase ID'].count()*100)/(unique['SN'].count()),2)).astype(str)+'%'})
display(age_demographics)
```

```python
age2=df.groupby(pd.cut(df['Age'],bins,labels=bin_names)) #We want to bin the original data, not unique with duplicate names dropped
age_analysis=pd.DataFrame({'Purchase Count':age2['Purchase ID'].count(),'Average Purchase Price':round(age2['Price'].mean(),2),'Total Purchase Value':age2['Price'].sum(),'Average Total Purchase per Person':(age2['Price'].sum()/age2['SN'].nunique())})
with pd.option_context('display.float_format', '${:,.2f}'.format):
    display(age_analysis)
```

```python
spenders=df.groupby(df['SN'])
top_spenders=pd.DataFrame({'Purchase Count':spenders['Purchase ID'].count(),'Average Purchase Price':spenders['Price'].mean(),'Total Purchase Value':spenders['Price'].sum()})
top_spenders=top_spenders.sort_values('Total Purchase Value',ascending=False)
with pd.option_context('display.float_format', '${:,.2f}'.format):
    display(top_spenders.head())
```

```python
items=df.groupby(['Item ID','Item Name'])
popular_items=pd.DataFrame({'Purchase Count':items['Purchase ID'].count(),'Item Price':items['Price'].mean(),'Total Purchase Value':items['Price'].sum()})
popular_items=popular_items.sort_values('Purchase Count', ascending=False)
with pd.option_context('display.float_format', '${:,.2f}'.format):
    display(popular_items.head())
```

```python
profitable_items=popular_items.sort_values('Total Purchase Value', ascending=False)
with pd.option_context('display.float_format', '${:,.2f}'.format):
    display(profitable_items.head())
```
